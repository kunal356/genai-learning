```python
from langchain.chains import LLMChain
from langchain import PromptTemplate

generictemplate = """
    Write a summary of the following speech:
    Speech: {speech}
    Translate the precise summary to {language}
"""

prompt = PromptTemplate(
    input_variable=['speech', 'language'],
    template=generictemplate
)
prompt
```

```python
complete_prompt = prompt.format(speech=speech, language="French")
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain = LLMChain(prompt=prompt, llm=llm)
summary = llm_chain.run({"speech": speech, "language": "Marathi"})
summary
```

### StuffDocumentChain Text Summarization

```python
from langchain.document_loaders import PyPDFLoader

loader = PyPDFLoader('apjspeech.pdf')
docs = loader.load_and_split()
docs
```

```python
template = """ Write a concise and short summary of the following Speech : {text}
"""
prompt = PromptTemplate(input_variable = ['text'],
                        template=template)
```

```python
from langchain.chains.summarize import load_summarize_chain
```

```python
chain = load_summarize_chain(llm, chain_type='stuff', verbose=True, prompt=prompt)
output_summary=chain.run(docs)
output_summary
```

### Map reduce summarize

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
docs
```

```python
final_documents = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=200).split_documents(docs)
final_documents
```

```python
len(final_documents)
```

```python
chunks_prompt="""
Please summarize the below speech:
Speech: `{text}`
Summary:
"""

map_prompt_template = PromptTemplate(input_variables=['text'],
                                     template=chunks_prompt)
```

```python
final_prompt="""
Provide the final summary of the entire speech with these important points.
Add a Motivation title, start the precise summary with an introduction and provide summary in number
points for the speech.
Speech: {text}
"""
final_prompt_template = PromptTemplate(input_variables=['text'], template=final_prompt)
```

```python
summary_chain = load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt=map_prompt_template,
    combine_prompt=final_prompt_template,
    verbose=True
)

output = summary_chain.run(final_documents)
output
```

### Refine Chain Summarization

```python
chain = load_summarize_chain(
    llm=llm,
    chain_type="refine",
    verbose=True
)

output_summary=chain.run(final_documents)
output_summary
```
