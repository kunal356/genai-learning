#### Chroma
Chroma is a AI-native open-source vector database focused on developer productivity and happiness. Chroma is licensed under Apache 2.0.

https://python.langchain.com/v0.2/docs/integrations/vectorstores/

```python
from langchain_chroma import Chroma
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import OllamaEmbeddings
from langchain.document_loaders import TextLoader
```

```python
loader = TextLoader("speech.txt")
documents = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=30)
docs = text_splitter.split_documents(documents)
```

```python
docs
```

```python
embeddings = OllamaEmbeddings(model='mxbai-embed-large')
db=Chroma.from_documents(docs, embeddings)
```

```python
## query it
query = "What does the speaker believe is the main reason the United States should enter the war?"
res = db.similarity_search(query=query)
res
```

```python
res_with_score = db.similarity_search_with_score(query)
print(res_with_score)
```

```python
embedding_vectors = embeddings.embed_query(query)
embedding_vectors
```

```python
vector_res = db.similarity_search_by_vector(embedding_vectors)
vector_res
```

#### Saving to the disc

```python
db = Chroma.from_documents(documents= docs, embedding= embeddings, persist_directory="./chroma_db")
```

```python
new_db = Chroma(persist_directory='./chroma_db',embedding_function=embeddings)
docs = new_db.similarity_search(query)
print(docs[0].page_content)
```

```python
res_with_score = new_db.similarity_search_with_score(docs)
print(res_with_score)
```
